```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_theme()
```

```python

df=pd.read_csv("marketing_campaign.csv" , sep = '\t')

```

```python
# task 2 part 3
numeric_df = df.select_dtypes(include=['float64', 'int64'])

sns.set(font_scale=0.6)
plt.figure(figsize=(14, 12))
sns.heatmap(numeric_df.corr(),annot=True)
```

```python

# task 2 
sns.boxplot(x='Income', data = numeric_df)  
```

```python
# task 2 part 1
sns.histplot(x = 'Income',kde = True , data = numeric_df) #overlay kde
plt.ylabel('frequency')
```

```python
sns.jointplot( x= 'Response', y= 'Income', kind = 'scatter', data = numeric_df , alpha=0.5 , color='g')
```
