```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('heart.csv')
plt.figure(figsize=(14, 20))
plt.subplot(8, 2, 1)
plt.grid(True)
sns.histplot(data, x='age', color='lightblue', kde=True)
plt.title('Age Distribution')
plt.subplot(8, 2, 2)
plt.grid(True)
sns.countplot(data=data, x='sex', palette='pastel')
plt.title('Gender Distribution')
plt.subplot(8, 2, 3)
plt.grid(True)
sns.countplot(data=data, x='cp', palette='viridis')
plt.title('Chest Pain Type Distribution')
plt.subplot(8, 2, 4)
plt.grid(True)
sns.histplot(data, x='trestbps', color='salmon', kde=True)
plt.title('Resting Blood Pressure Distribution')
plt.subplot(8, 2, 5)
plt.grid(True)
sns.histplot(data, x='chol', color='orange', kde=True)
plt.title('Cholesterol Level Distribution')
plt.subplot(8, 2, 6)
plt.grid(True)
sns.countplot(data=data, x='fbs', palette='coolwarm')
plt.title('Fasting Blood Sugar Levels')
plt.subplot(8, 2, 7)
plt.grid(True)
sns.countplot(data=data, x='restecg', palette='Set1')
plt.title('Resting ECG Results')
plt.subplot(8, 2, 8)
plt.grid(True)
sns.histplot(data, x='thalach', color='purple', kde=True)
plt.title('Maximum Heart Rate Distribution')
plt.subplot(8, 2, 9)
plt.grid(True)
sns.countplot(data=data, x='exang', palette='Blues')
plt.title('Exercise Induced Angina')
plt.subplot(8, 2, 10)
plt.grid(True)
sns.histplot(data, x='oldpeak', color='red', kde=True)
plt.title('Oldpeak (ST Depression) Distribution')
plt.subplot(8, 2, 11)
plt.grid(True)
sns.countplot(data=data, x='slope', palette='YlGnBu')
plt.title('Slope of Peak Exercise ST Segment')
plt.subplot(8, 2, 12)
plt.grid(True)
sns.countplot(data=data, x='ca', palette='coolwarm')
plt.title('Number of Major Vessels Colored')
plt.subplot(8, 2, 13)
plt.grid(True)
sns.countplot(data=data, x='thal', palette='magma')
plt.title('Thalassemia')
plt.subplot(8, 2, 14)
plt.grid(True)
sns.countplot(data=data, x='target', palette='crest')
plt.title('Heart Disease Diagnosis')
plt.tight_layout()
plt.show()

# ------------------- Paper 2 -------------------
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x="fbs", kde=False, color='seagreen')
plt.title('Distribution of Fasting Blood Sugar (fbs)')
plt.show()
plt.figure(figsize=(8, 6))
sns.boxplot(x="fbs", y="age", data=data, palette='muted')
plt.title('Boxplot of Age vs. Fasting Blood Sugar Levels')
plt.show()
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x="fbs", y="age", hue='target', palette='coolwarm', s=100, edgecolor='black')
plt.title('Scatter Plot of Age vs. Fasting Blood Sugar (fbs)')
plt.show()
plt.figure(figsize=(8, 6))
sns.lineplot(x="fbs", y="age", data=data, marker='o', color='salmon')
plt.title('Line Plot of Age vs. Fasting Blood Sugar (fbs)')
plt.show()
plt.figure(figsize=(10, 8))
sns.displot(data, col="fbs", kde=True)
plt.ylabel("FBS Level")
plt.show()
sns.pairplot(data, hue="target", palette='dark:blue', markers=["o", "s"], diag_kind="kde")
plt.title('Pairplot of Dataset with Heart Disease Diagnosis')
plt.show()
```
